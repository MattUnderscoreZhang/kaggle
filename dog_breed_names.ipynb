```python
import pandas as pd
```

```python
train = pd.read_csv("data/train.csv")
train = train[train.AnimalType=="Dog"]
```

```python
detail = pd.read_json("breeds/dog_detail.json")
detail = detail.apply( lambda x: pd.Series( x.values[0] ),axis=1 )
```

```python
breeds = train.Breed.apply(lambda x:x.replace("Mix",""))
```

```python
def breed_after_slash(entry):
    if "/" in entry:
        return entry.split("/")[1]
    else:
        return entry.strip(" ")
    # end if
# end def
```

```python
import re
def dog_breed_info_name(name):
    if "Terr" in name:
        # replace full word, \b defines word boundary
        return re.sub(r'\bTerr\b','',name)
    elif "English Bulldog" in name:
        return "Bulldog"
    elif name == "Anatol Shepherd":
        return "Anatolian Shepherd Dog"
    elif "Yorkshire" in name:
        return "Yorkshire Terrier"
    elif "Pit Bull" in name:
        return "American Pit Bull Terrier"
    elif name == "Wire Hair Fox Terrier":
        return "Wire Fox Terrier"
    elif name == "Chihuahua Shorthair":
        return "Chihuahua"
    elif name == "Dachshund":
        return "Dachshund (Standard)"
    else:
        return name
    # end if
# end def
```

```python
test = breeds.apply(breed_after_slash).apply(dog_breed_info_name)
```

```python
def fix_dog_breed_name(name):
    if "Rat Terrier" in name:
        return "Rat Terrier"
    else:
        return name
    # end if
# end def
```

```python
detail["Name"] = detail.Name.apply(fix_dog_breed_name)
```

```python
num_match = 0
for breed in test.values:
    if breed in detail.Name.values:
        num_match += 1
    else:
        print breed
    # end if
# end for
float( num_match )/len(test)
```

```python
known_breeds = set()
for breed in breeds:
    myname = ""
    if "/" in breed:
        for name in breed.split("/"):
            myname = name.strip(" ")
        # end if
    else:
        myname = breed.strip(" ") 
    # end if
    known_breeds.add( dog_breed_info_name(myname) )
# end for
```

```python
has_data = 0
for breed in known_breeds:
    if breed in detail.Name.values:
        has_data += 1
    else:
        print breed
    # end if
# end for
float(has_data)/len(known_breeds)
```

```python
def sort_sizes(entry):
    types = ["small","medium","large","giant"]
    for mytype in types:
        if mytype in entry.lower():
            return mytype
        # end if
    # end for
    
    # should have returned at this point
    # --- deal with special cases below
    
    if "Butterfly" in entry or "Squirrel" in entry:
        return "small"
    # end if
    elif "Samll" in entry:
        return "small"
    elif "Working" in entry:
        return "giant"
    else:
        return "other"
    # end if
    
    return entry
# end def
```

```python
detail["body_size"] = detail.Type.apply(sort_sizes)
```
